# 데이터 전처리

## 메타데이터 이용해 암페어 변수 추가

* 메타데이터 -> 데이터 프레임으로 정리
  * GI 외 + GI 오일링 (품명그룹 별) 두 시트로 이루어져있으니 각각 다르게 정리
  * 두께 그룹이 '1.9t 이상' 을 '1.9≤t＜5.0'로 고쳐서 train 과 동일하도록.
  * 두께그룹이 0.60으로 들어가있어 매핑 안돼 0.6으로 소수점 한자리로
  * train/test에는 두께그룹이 0.6≤t＜0.7, metadata에서는 0.6＜t＜0.7, train에 맞게 맞추기

```python
# 메타데이터 엑셀 파일의 시트 이름 확인
metadata_path = '/content/drive/MyDrive/ToBigs/2025-mix-up-ai-datathon-track2/track2_metadata.xlsx'
xls = pd.ExcelFile(metadata_path)
sheet_names = xls.sheet_names

# 두 시트 로드
metadata_gi = pd.read_excel(metadata_path, sheet_name=0)
metadata_else = pd.read_excel(metadata_path, sheet_name=1)
```

```python
import re

def standardize_group_format(s):
    if pd.isna(s):
        return s
    s = s.replace('1.9t 이상', '1.9≤t＜5.0') # '1.9t 이상' → '1.9≤t＜5.0' 문자열로 바꾸기
    s = re.sub(r'([\d]+\.[\d]+)', lambda m: f'{float(m.group(1)):.1f}', s)  # 소수점 1자리 0.50 -> 0.5
    s = s.replace('0.6＜t＜0.7', '0.6≤t＜0.7')
    return s
```

```python
# 1. GI 오일링 기준표 정리 (두께그룹, 폭그룹, Ampere 만)
metadata_gi_long = metadata_gi.iloc[3:, :12].copy()
metadata_gi_long.columns = ['두께그룹', '800≤w＜900', '900≤w＜1000', '1000≤w＜1100',
                            '1100≤w＜1200', '1200≤w＜1300', '1300≤w＜1400', '1400≤w＜1500',
                            '1500≤w＜1600', '1600≤w＜1700', '1700≤w＜1800', '1800≤w≤']
metadata_gi_long = metadata_gi_long[metadata_gi_long['두께그룹'].notnull()]
metadata_gi_long = metadata_gi_long.melt(id_vars='두께그룹', var_name='폭그룹', value_name='기준Ampere')
metadata_gi_long['두께그룹'] = metadata_gi_long['두께그룹'].apply(standardize_group_format)
metadata_gi_long = metadata_gi_long[['두께그룹', '폭그룹', '기준Ampere']]

# 2. GI 외 기준표 정리
metadata_else_long = metadata_else.iloc[3:, :12].copy()
metadata_else_long.columns = ['두께그룹', '800≤w＜900', '900≤w＜1000', '1000≤w＜1100',
                              '1100≤w＜1200', '1200≤w＜1300', '1300≤w＜1400', '1400≤w＜1500',
                              '1500≤w＜1600', '1600≤w＜1700', '1700≤w＜1800', '1800≤w≤1850']
metadata_else_long = metadata_else_long[metadata_else_long['두께그룹'].notnull()]
metadata_else_long = metadata_else_long.melt(id_vars='두께그룹', var_name='폭그룹', value_name='기준Ampere')
metadata_else_long['두께그룹'] = metadata_else_long['두께그룹'].apply(standardize_group_format)
metadata_else_long = metadata_else_long[['두께그룹', '폭그룹', '기준Ampere']]
```

```python
# 기준Ampere merge
def merge_ampere(df, metadata_gi, metadata_else):
    df_gi = df[df['품명그룹'].str.contains('오일링', na=False)].merge(
        metadata_gi, on=['두께그룹', '폭그룹'], how='left'
    )
    df_else = df[df['품명그룹'].str.contains('외', na=False)].merge(
        metadata_else, on=['두께그룹', '폭그룹'], how='left'
    )
    return pd.concat([df_gi, df_else], ignore_index=True)

# train/test merge
train = merge_ampere(train, metadata_gi_long, metadata_else_long)
test = merge_ampere(test, metadata_gi_long, metadata_else_long)
```

## 결측치 처리

강종, 소재길이, 도유, 장력1, 스피드1, 장력2, 스피드2, 장력3, 스피드3

```python
import matplotlib.pyplot as plt

# 결측치 비율 계산 함수
def missing_values_table(df):
    mis_val = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis_val_table_ren_columns = mis_val_table.rename(
        columns={0: 'Missing Values', 1: '% of Total Values'})
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:, 1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
    return mis_val_table_ren_columns

# train 결측치 비율
missing_values_table(train)
```

* 장력변수들, 스피드변수들, 소재길이 -> 0으로 대체
* 강종, 도유 (범주형) -> 'Unknown'으로

```python
# 범주형 변수 결측 처리
cat_fill_cols = ['도유', '강종']
for col in cat_fill_cols:
    train[col] = train[col].fillna('Unknown')
    test[col] = test[col].fillna('Unknown')

# 수치형 결측 처리
num_fill_cols = ['소재길이', '스피드1', '장력1', '스피드2', '장력2', '장력3', '스피드3']
for col in num_fill_cols:
    train[col] = train[col].fillna(0)
    test[col] = test[col].fillna(0)
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, roc_auc_score

# 대상 변수
force_cols = ['장력1', '장력2', '장력3', '장력']
speed_cols = ['스피드1', '스피드2', '스피드3', '스피드']
std_force_cols = [f'{col}_std' for col in force_cols]

# ✔️ 정확히 제외 (장력, 스피드, 표준화 모두 제외)
other_cols = [col for col in train.columns if col not in ['OK', 'NG'] + force_cols + speed_cols + std_force_cols]

y = train['NG']

# ✔️ 1. 전체 원래 변수 사용 (장력, 스피드 포함)
X1 = train[other_cols + force_cols + speed_cols]

# ✔️ 2. 전체 원래 변수 + 표준화 변수 (장력, 스피드 + 표준화)
X2 = train[other_cols + force_cols + speed_cols + std_force_cols]

# ✔️ 3. 원래 변수에서 장력1,2,3,장력 제거하고 표준화 변수만 추가
X3 = train[other_cols + speed_cols + std_force_cols]

# ✔️ 함수로 공통 평가
def evaluate_model(X, y, desc):
    X_train, X_val, y_train, y_val = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
    model = LGBMClassifier(class_weight='balanced', random_state=42)
    model.fit(X_train, y_train)
    f1 = f1_score(y_val, model.predict(X_val))
    roc_auc = roc_auc_score(y_val, model.predict_proba(X_val)[:, 1])
    final_score = (f1*0.4 + roc_auc*0.6)**2
    print(f"[{desc}] F1: {f1:.4f}, ROC-AUC: {roc_auc:.4f}, Final Socre: {final_score:.4f}")

# ✔️ 평가 실행
evaluate_model(X1, y, '원래 변수 (장력1,2,3,장력 포함)')
evaluate_model(X2, y, '원래 변수 + 표준화')
evaluate_model(X3, y, '원래 변수에서 장력1,2,3,장력 제거 + 표준화')
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score, roc_auc_score
from lightgbm import LGBMClassifier

# 모델링 (categorical_feature 직접 지정, 인코딩 X)
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
oof_preds = np.zeros(len(y))
test_preds = np.zeros(len(test))

for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
    print(f'Fold {fold + 1}')
    X_tr, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = LGBMClassifier(random_state=42, class_weight='balanced', n_estimators=1000, learning_rate=0.03)
    model.fit(X_tr, y_tr,
              eval_set=[(X_val, y_val)],
              categorical_feature=cat_cols)

    val_prob = model.predict_proba(X_val)[:, 1]
    oof_preds[val_idx] = val_prob
    test_preds += model.predict_proba(test)[:, 1] / kf.n_splits

# -----------------------
# 평가
f1 = f1_score(y, (oof_preds >= 0.5).astype(int))
roc_auc = roc_auc_score(y, oof_preds)
final_score = (f1 * 0.4 + roc_auc * 0.6) ** 2

print(f'\n✅ OOF F1: {f1:.4f}, ROC-AUC: {roc_auc:.4f}, Final Score: {final_score:.4f}')
```

```python
# ✅ LightGBM, CatBoost, XGBoost, VotingClassifier 자동 비교 + 최적 모델 추천 + 제출 파일 생성
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score, roc_auc_score
from sklearn.ensemble import VotingClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier

# Label Encoding

X_encoded = X.copy()
test_encoded = test.copy()

for col in cat_cols:
    le = LabelEncoder()
    all_values = pd.concat([X_encoded[col], test_encoded[col]], axis=0).astype(str).fillna('missing')
    le.fit(all_values)
    X_encoded[col] = le.transform(X_encoded[col].astype(str).fillna('missing'))
    test_encoded[col] = le.transform(test_encoded[col].astype(str).fillna('missing'))


# 모델 정의
lgbm = LGBMClassifier(random_state=42, class_weight='balanced', n_estimators=1000, learning_rate=0.03)
cat = CatBoostClassifier(random_state=42, iterations=1000, learning_rate=0.03, verbose=0, auto_class_weights='Balanced')
xgb = XGBClassifier(random_state=42, n_estimators=1000, learning_rate=0.03, use_label_encoder=False, eval_metric='logloss', scale_pos_weight=1.5)
ensemble = VotingClassifier(estimators=[('lgbm', lgbm), ('cat', cat), ('xgb', xgb)], voting='soft')

models = {
    'LightGBM': lgbm,
    'CatBoost': cat,
    'XGBoost': xgb,
    'Voting': ensemble
}

results = {}

# 모델별 비교
for name, model in models.items():
    print(f'\n====== {name} START ======')
    oof_preds = np.zeros(len(y))
    test_preds = np.zeros(len(test))

    kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    for fold, (train_idx, val_idx) in enumerate(kf.split(X_encoded, y)):
        print(f'{name} Fold {fold + 1}')
        X_tr, X_val = X_encoded.iloc[train_idx], X_encoded.iloc[val_idx]
        y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]

        model.fit(X_tr, y_tr)
        val_prob = model.predict_proba(X_val)[:, 1]
        oof_preds[val_idx] = val_prob
        test_preds += model.predict_proba(test_encoded)[:, 1] / kf.n_splits

    f1 = f1_score(y, (oof_preds >= 0.5).astype(int))
    roc_auc = roc_auc_score(y, oof_preds)
    final_score = (f1 * 0.4 + roc_auc * 0.6) ** 2

    print(f'{name} F1: {f1:.4f}, ROC-AUC: {roc_auc:.4f}, Final Score: {final_score:.4f}')
    results[name] = {'f1': f1, 'roc_auc': roc_auc, 'final_score': final_score, 'test_preds': test_preds}

# 최적 모델 선정
best_model_name = max(results, key=lambda x: results[x]['final_score'])
print(f'\n✅ 최종 선택 모델: {best_model_name} | Final Score: {results[best_model_name]["final_score"]:.4f}')
```
